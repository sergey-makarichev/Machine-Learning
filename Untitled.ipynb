# 0. Описание задачи
## В датасете имеются следующие признаки:
- Age - возраст
- Job - профессия
- Marital Status - семейное положение
- Education - образование
- Credit - кредитное страхование
- Balance (euros) - среднегодовой баланс
- Housing Loan - наличие жилищного кредита
- Personal Loan - наличие обычного кредита
- Contact - способ связи с клиентом
- Last Contact Day - день последней связи с клиентом
- Last Contact Month - месяц последней связи с клиентом
- Last Contact Duration - продолжительность последнего контакта представителя банка с клиентом
- Campaign - количестов контактов с клиентом
- Pdays - количество дней, прошедших после того, как с клиентом в последний раз связывались из предыдущей кампании (числовое значение -1 означает, что с клиентом ранее не связывались)
- Previous - количество контактов, выполненных до этой кампании с данным клиентом
- Poutcome - итоги предыдущей маркетинговой кампании
- Subscription - ключевое знаечние - оформил ли клиент срочный депозит?  
Задача связана с маркетинговыми кампаниями португальского банковского учреждения. Маркетинговые кампании были основаны на телефонных звонках. Часто требовалось более одного контакта с одним и тем же клиентом, чтобы получить ответ, будет ли  подписан банковский продукт (срочный депозит).  
Нужно на основе данных признаков предсказать, будет ли клиентом принято предложение банка.

# 4. Обработка категориальных признаков

Бинарные признаки `'Credit'`, `'Housing Loan'` `'Personal Loan'` `'Subscription'` заменим на числовые. 

```python
data['Credit'] = pd.factorize(data['Credit'])[0]
data['Housing Loan'] = pd.factorize(data['Housing Loan'])[0]
data['Personal Loan'] = pd.factorize(data['Personal Loan'])[0]
data['Subscription'] = pd.factorize(data['Subscription'])[0]
```

К категориальным признакам `'Job'`, `'Marital Status'`, `'Education'`, `'Contact Last'`, `'Contact Month'`, `'Poutcome'` применим метод бинаризации

```python
Job_dummies = pd.get_dummies(data['Job'])
Marital_Status_dummies = pd.get_dummies(data['Marital Status'])
Education_dummies = pd.get_dummies(data['Education'])
Contact_dummies = pd.get_dummies(data['Contact'])
Last_Contact_Month_dummies = pd.get_dummies(data['Last Contact Month'])
Poutcome_dummies = pd.get_dummies(data['Poutcome'])
```

Добавим эти dummy-столбцы к таблице и удалим столбцы `'Job'`, `'Marital Status'`, `'Education'`, `'Contact'`, `'Last Contact Month'`, `'Poutcome'`:

```python
data = pd.concat((data, Job_dummies, Marital_Status_dummies, Education_dummies, Contact_dummies, Last_Contact_Month_dummies, Poutcome_dummies), axis=1)
data = data.drop(['Job', 'Marital Status', 'Education', 'Contact', 'Last Contact Month', 'Poutcome'], axis=1)
```

# 6. Разбиение данных на обучающую и тестовую выборку

```python
X = data.drop('Subscription', axis = 1) # выбрасываем столбец 'Subscription'
y = data['Subscription'] # создаём вектор значений столбца 'Subscription'
feature_names = X.columns
```

```python
feature_names
```

```python
X = X.to_numpy()
```

```python
Y = np.array(y)
type(Y)
```

```python
print(X.shape)
print(y.shape)
N, d = X.shape
```

```python
Y
```

Разбиваем множества на обучающую и тестовую части

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

# 7. Запускаем классификатор ближайших соседей

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)
```

```python
knn
```

### Попробуем предсказать значение, используя функцию  predict

```python
number = 56
obj = X_test[number, :]
value = y_test[number]
knn.predict([obj])
```

```python
value
```

# 8. Вычисление ошибок на обучащей и тестовой выборках

```python
y_train_pred = knn.predict(X_train)
y_test_pred = knn.predict(X_test)
```

```python
y_train != y_train_pred
```

```python
y_test != y_test_pred
```

Находим долю ошибок в нашем предсказание

```python
y_test_pred = knn.predict(X_test)
err_test  = np.mean(y_test  != y_test_pred)
```

```python
y_train_pred = knn.predict(X_train)
err_train  = np.mean(y_train  != y_train_pred)
```

```python
err_test, err_train
```

Находим качество нашего предсказания

```python
knn.score(X_train, y_train), knn.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_pred))
```

```python
confusion_matrix([0,1,0,1,0], [0,1,0,0,0])
```

Ошибка на тестовой выборке больше ошибки на обучающей выборке. Это является проявлением переобучения.

# 9. Сравнение с другими методами

## Метод SVC

```python
from sklearn.svm import SVC
svc = SVC(gamma='auto')
svc.fit(X_train, y_train)

err_train = np.mean(y_train != svc.predict(X_train))
err_test  = np.mean(y_test  != svc.predict(X_test))

print(err_train, err_test)
```

У данного метода намного меньше ошибка на обучающей выборке и немного меньше на тестовой по сравнению с предыдущим. Но скорость работы у метода SVC ниже.
